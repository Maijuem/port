```python
import numpy as np
import pandas as pd
```

```python
# import data from csv file
covid19=pd.read_csv('full_group.csv')
covid19.head()
```

```python
# Check the shape of data frame and NaN value
## It seems that there is no NA value in the data frame
covid19.shape
pd.isnull(covid19)
```

```python
# Sort by two columns, first is by month, second value is by day
## As we can see from the table, earliest date is 22-01-2020 and the latest day is 27-07-2020
covid19_sorted= covid19.sort_values(by=['month', 'day'])
covid19_sorted
```

```python
print('Earliest date appears on the table is :  ', covid19.Date.min())
```

```python
print('Latest date appears on the table is:  ', covid19.Date.max())
```

```python
#select and print out April data
april_data=covid19[covid19.month==4]
april_data
```

```python
# Determine number of death in this month by taking the sum of column New_deaths
april_data.New_deaths.sum()
```

```python
#Create new frame with data from 2 column CountryRegion and WHO_region
## using iloc
## print out the new frame
new_frame= covid19.iloc[:, [1,9]]
new_frame
```

```python
#Remove duplicate so that each country in 1 region occurs only once
singlecountry=new_frame.drop_duplicates()
singlecountry
```

```python
#Group country by WHO region to determine which region has the most countries associate to it
## according to below table, we can see that Africa has the biggest number of countries.
country_by_region= singlecountry.groupby(['WHO_Region'])
country_by_region.size()
```

```python
#select data for USA only
us_data=covid19[covid19.CountryRegion=='US']

#create a month list that contains April and March then filter data for USA in only these month
monthlist=[3,4]
march_and_april = covid19[(covid19.CountryRegion== 'US') & (covid19.month.isin(monthlist))]
march_and_april
```

```python
if us_data[us_data.month==3].New_deaths.mean()> us_data[us_data.month==4].New_deaths.mean():
    print('Average number of daily new death in March in US was higher')
else:
    print('Average number of daily new death in April in US was higher')
```

```python
# Check again in detail data of the USA in March
US_march= us_data[us_data.month==3]
US_march.describe()
```
